This Notebook was purely used for statistical testing.  

Based on the Friedman test results both 0 hypotheses (no difference in embeddings and no differences in algortithms in sensitivity to MT induced noise) were rejected.  
Nemenyi post-hoc tests were performed to pinpoint which embeddings actually differed from each other, and the same was done for the algortihms.  

\\

*More details about the design be found on this repo's readme here: [README](https://github.com/StevenPeutz/Disinformation-NLP/blob/master/README.md) , and the main code can be found here: [MAIN](https://github.com/StevenPeutz/Disinformation-NLP/blob/master/MAIN.ipynb).*

\\

---


*(No manual correction for multiple comparisons was performed, as this correction is built in to the Nemenyi test itself)*.

```python
data_bow = [-0.4189071632,	-0.2826967514,	-0.380909338,	-0.4254315618,	-0.2441824113,	-0.4881337502,	-0.448769879]
data_hv = [-0.2871075167,	-0.3275927688,	-0.2952222373,	-0.4369104254,	-0.1747587332,	-0.4403901047,	-0.4123963572]
data_tfidf = [-0.3142041593,	-0.3157469077,	-0.23889493,	-0.23331521,	-0.1966698382,	-0.4413415794,	-0.4397036836]
data_glove = [-0.2674255811,	-0.136461873,	-0.3638575506,	-0.1433940465,	-0.1765393503,	-0.396690227,	-0.3624643197]
data_w2v = [-0.3547981514,	-0.2057768112,	-0.4155498165,	-0.4408862308,	-0.3311185299,	-0.4372706266,	-0.4555593312]
data_ft = [-0.2683498709,	-0.1431018078,	-0.457740927,	-0.3350754384,	-0.1663245888,	-0.3973018894,	-0.4080059807]
```

```python
data_lr = [-0.4189071632, -0.2871075167, -0.3142041593, -0.2674255811, -0.3547981514, -0.2683498709]
data_nb = [-0.2826967514, -0.3275927688, -0.3157469077, -0.136461873, -0.2057768112, -0.1431018078]
data_rf = [-0.380909338, -0.2952222373, -0.23889493, -0.3638575506, -0.4155498165, -0.457740927]
data_svm = [-0.4254315618, -0.4369104254, -0.23331521, -0.1433940465, -0.4408862308, -0.3350754384]
data_knn = [-0.2441824113, -0.1747587332, -0.1966698382, -0.1765393503, -0.3311185299, -0.1663245888] #better to remove...
data_gb = [-0.4881337502, -0.4403901047, -0.4413415794, -0.396690227, -0.4372706266, -0.3973018894]
data_xgb = [-0.448769879, -0.4123963572, -0.4397036836, -0.3624643197, -0.4555593312, -0.4080059807]
```

# Friedman tests (algs & embeds)

```python
#friedman test algs
stats.friedmanchisquare(data_lr, data_nb, data_rf, data_svm, data_knn, data_gb, data_xgb)
```

```python
#friedman test embeddings
stats.friedmanchisquare(data_bow, data_hv, data_tfidf, data_glove, data_w2v, data_ft)
#pvalue=0.0054 , when running withhout the knn mesurements, it is still significant (at pvalue=0.0133)
```

# posthoc -> algorithms

```python
#post hoc algs

#combine three groups into one array
data_algs = np.array([data_lr, data_nb, data_rf, data_svm, data_knn, data_gb, data_xgb])

#perform Nemenyi post-hoc test
sp.posthoc_nemenyi_friedman(data_algs.T)

# 1 vs 5: naive bayes vs gradient boosting
# 4 vs 5: knn vs vs gradient boosting
# 4 vs 6: knn vs xgb
```

```python
#post hoc algs (no knn)


#combine three groups into one array
data_algs_noknn = np.array([data_lr, data_nb, data_rf, data_svm, data_gb, data_xgb])

#perform Nemenyi post-hoc test
sp.posthoc_nemenyi_friedman(data_algs_noknn.T)

# 1 vs 4: naive bayes vs gradient boosting
```

# posthoc -> embeddings


```python
#post hoc embeddings
#data_bow, data_hv, data_tfidf, data_glove, data_w2v, data_ft

#combine three groups into one array
data_embeds = np.array([data_bow, data_hv, data_tfidf, data_glove, data_w2v, data_ft])

#perform Nemenyi post-hoc test
sp.posthoc_nemenyi_friedman(data_embeds.T)

#0 vs 3: bow vs glove
#3 vs 4: glove vs w2v
```
